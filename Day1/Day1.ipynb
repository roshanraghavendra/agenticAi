```python
# Stemming - extract root word
from nltk.stem import PorterStemmer
stemmer_obj=PorterStemmer()
stemmer_obj.stem("going")
# package  - nltk
 #  |--subpackage -stem
 #       |------class PorterStemmer
 #                ..def stem()

stemmer_obj.stem("eating")
```

```python
# Regular Expression
from nltk.stem import RegexpStemmer
reg_stemmer= RegexpStemmer('ing$|s$|es$')
print(reg_stemmer.stem('eating'))
print(reg_stemmer.stem('going'))
print(reg_stemmer.stem('running'))
```
